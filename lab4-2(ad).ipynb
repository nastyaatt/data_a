Конвертуємо файл

```python
import csv

def to_csv():
    txt_file = 'D:/anaconda3/envs/lab4/hepatitis_data.txt'
    csv_file = 'hepatitis_data.csv'

    with open(txt_file, 'r') as file_in, open(csv_file, 'w', newline='') as file_out:
        reader = csv.reader(file_in, delimiter=',')
        writer = csv.writer(file_out)

        for row in reader:
            writer.writerow(row)
    print(f"Успішно конвертовано {txt_file} у {csv_file}")
to_csv()
```

```python
import numpy as np

def np_df(file):
    data = np.genfromtxt(file, delimiter=',', dtype=str, skip_header=1)
    data = data[~np.any(data == '?', axis=1)]

    data[:, 0] = np.where(data[:, 0] == '1', 'die', 'live')
    data[:, 2] = np.where(data[:, 2] == '1', 'male', 'female') 
    data[:, 3:14] = np.where(data[:, 3:14] == '1', 'no', 'yes') 
    data[:, -1] = np.where(data[:, -1] == '1', 'no', 'yes')  
    
    numeric_columns = [1, 14, 15, 16, 17, 18]
    data[:, numeric_columns] = data[:, numeric_columns].astype(float)
    
    return data

file = "hepatitis_data_np.csv"
data = np_df(file)
print(data[:30])
```

2. Пронормувати вибраний датасет або стандартизувати його. (Обрала нормалізацію) (Pandas)

```python
def normalize_data(df):
    for column in df.columns:
        if df[column].dtype == 'float64' or df[column].dtype == 'int64':
            df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())
    return df
normalize_data(df)
df = normalize_data(df)
print(df.head())

```

2. Пронормувати вибраний датасет або стандартизувати його. (Обрала нормалізацію) (NumPy)

```python
def normalize_data_np(data, numeric_columns):
    for col_idx in numeric_columns:
        col_min = np.min(data[:, col_idx].astype(float))
        col_max = np.max(data[:, col_idx].astype(float))
        if col_max != col_min:
            data[:, col_idx] = (data[:, col_idx].astype(float) - col_min) / (col_max - col_min)
    return data

numeric_columns = [1, 14, 15, 16, 17, 18]
data = normalize_data_np(data, numeric_columns)
print(data[:10])
```

5. Підрахувати коефіцієнт Пірсона та Спірмена для двох integer/real атрибутів.(Pandas)

```python
from scipy.stats import pearsonr, spearmanr

def pir_pd(df):
    x = df['Bilirubin']
    y = df['Albumin']

    pearson_corr, pearson_p_value = pearsonr(x, y)
    spearman_corr, spearman_p_value = spearmanr(x, y)
    print("Коефіцієнт Пірсона:", pearson_corr)
    print("Коефіцієнт Спірмена:", spearman_corr)
pir_pd(df)
```

6. Провести One Hot Encoding категоріального string атрибуту.(NumPy)

```python
def one_hot_encode(data, column_index):
    unique_values = np.unique(data[:, column_index])
    
    encoded_columns = np.zeros((len(data), len(unique_values)), dtype=int)
    
    for i, value in enumerate(unique_values):
        encoded_columns[:, i] = (data[:, column_index] == value).astype(int)
    
    data = np.delete(data, column_index, axis=1)
    data = np.hstack((data, encoded_columns))
    
    return data

data = one_hot_encode(data, 2)
print(data[:5]) 
```
