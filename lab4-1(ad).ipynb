```python
import pandas as pd
import numpy as np
from datetime import time as dt_time

file = "household_power_consumption.csv"

def pd_df():

    file = "household_power_consumption.csv"
    dtype = {
        'Global_active_power': float,
        'Global_reactive_power': float,
        'Voltage': float,
        'Global_intensity': float,
        'Sub_metering_1': float,
        'Sub_metering_2': float,
        'Sub_metering_3': float
    }

    na_values = ['?']
    df = pd.read_csv(file, dtype=dtype, na_values=na_values)

    df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
    df["Time"] = pd.to_datetime(df["Time"], format='%H:%M:%S').dt.time
    df = df.drop(df.loc[(df["Sub_metering_1"] == 0) & (df["Sub_metering_2"] == 0) & (df["Sub_metering_3"] == 0)].index).dropna()
    return df
pd_df()

df = pd_df()

```

3. Обрати всі домогосподарства, у яких сила струму лежить в межах 19-20 А, для них виявити ті, у яких пральна машина та холодильних споживають більше, ніж бойлер та кондиціонер.

4. Обрати випадковим чином 500000 домогосподарств (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергії,

```python
def random_sample_average(df, n=500000):
    sampled_df = df.sample(n=n, replace=False)
    averages = sampled_df[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()
    return averages
random_sample_average(df, n=500000)
```

Обрати ті домогосподарства, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини.

```python
def complex_filter(df):
    filtered_df = df[(df['Time'].apply(lambda x: x > dt_time(hour=18))) & (df['Global_active_power'] > 6.0)]
    filtered_df = filtered_df[(filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_1']) &
                              (filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_3']) &
                              (filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_2'].mean())]
    first_half = filtered_df.head(len(filtered_df) // 2)
    second_half = filtered_df.tail(len(filtered_df) // 2)
    result = pd.concat([first_half.iloc[::3, :], second_half.iloc[::4, :]])
    return result
complex_filter(df)
```

4. Обрати випадковим чином 500000 домогосподарств (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергії,

```python
def random_sample_average_np(data, n=500000):
    sampled_indices = np.random.choice(data.shape[0], n, replace=False)

    sampled_data = data[sampled_indices, :]
    
    sub_metering_1 = sampled_data[:, 6].astype(float)
    sub_metering_2 = sampled_data[:, 7].astype(float)
    sub_metering_3 = sampled_data[:, 8].astype(float)
    
    averages = np.mean([sub_metering_1, sub_metering_2, sub_metering_3], axis=1)
    
    return averages

averages = random_sample_average_np(data, n=500000)
print(averages)
```

5. Обрати ті домогосподарства, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини.

```python
def complex_filter_np(data):
    time_column = 1
    power_column = 2
    sub_metering_1_column = 6
    sub_metering_2_column = 7
    sub_metering_3_column = 8

    times = np.array([t.split(':') for t in data[:, time_column]], dtype=int)
    after_18_indices = times[:, 0] >= 18
    filtered_data = data[after_18_indices]

    global_active_power = filtered_data[:, power_column].astype(float)
    high_power_indices = global_active_power > 6.0
    filtered_data = filtered_data[high_power_indices]

    sub_metering_1 = filtered_data[:, sub_metering_1_column].astype(float)
    sub_metering_2 = filtered_data[:, sub_metering_2_column].astype(float)
    sub_metering_3 = filtered_data[:, sub_metering_3_column].astype(float)

    dominant_group_2_indices = (sub_metering_2 > sub_metering_1) & (sub_metering_2 > sub_metering_3) & (sub_metering_2 > sub_metering_2.mean())
    filtered_data = filtered_data[dominant_group_2_indices]

    mid_index = len(filtered_data) // 2
    first_half = filtered_data[:mid_index]
    second_half = filtered_data[mid_index:]

    first_half_result = first_half[::3]
    second_half_result = second_half[::4]

    result = np.vstack((first_half_result, second_half_result))

    return result

filtered_data = complex_filter_np(data)
print(filtered_data)

```
